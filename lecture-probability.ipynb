```python
import numpy as np
import scipy.stats as scs
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
rv = scs.norm(15, 2)
```

```python
rv.stats('mvsk')
```

```python
xs = np.linspace(5, 25)
ys1 = rv.pdf(xs)
ys2 = rv.cdf(xs)
```

```python
fig, axes = plt.subplots(2, 1)
axes[0].plot(xs, ys1, label='pdf', color='red', alpha=0.3)
axes[1].plot(xs, ys2, label='cdf', color='green', alpha=0.5)
fig.legend();
```

```python
rv = scs.binom(10, 0.3)
xs = np.arange(12)
ys1 = rv.pmf(xs)
ys2 = rv.cdf(xs)
```

```python
fig, axes = plt.subplots(2, 1)
axes[0].bar(xs, ys1, label='pmf', color='blue', alpha=0.3)
axes[1].bar(xs, ys2, label='cdf', color='orange', alpha=0.5)
fig.legend();
```

```python
rv = scs.poisson(7)
rv.stats('mvsk')
```

```python
xs = np.arange(20)
ys1 = rv.pmf(xs)
ys2 = rv.cdf(xs)
```

```python
fig, axes = plt.subplots(2, 1)
axes[0].bar(xs, ys1, label='pmf', color='olive', alpha=0.3)
axes[1].bar(xs, ys2, label='cdf', color='yellow', alpha=0.5)
fig.legend();
```
